```python
from datos_pqr import cert_1427, reg_1427
import pandas as pd

class AnalisisCertificados:
    def __init__(self, cert_1427):
        self.datos = pd.DataFrame(cert_1427)
        self._preprocesar_datos()

    def _preprocesar_datos(self):
        # Dividir la columna 'Datos' en columnas separadas
        self.datos[['TD', 'Documento', 'Nombre1', 'Nombre2', 'Nombre3', 'Apellido1', 'Apellido2', 'Apellido3']] = self.datos['Datos'].str.split(' ', expand=True, n=7)

        # Filtrar valores nulos y duplicados en 'Documento'
        datos_fil = self.datos[['Documento']].copy()
        datos_fil['Documento'] = datos_fil['Documento'].replace('None', pd.NA)  # Reemplazar 'None' por NaN
        datos_fil.dropna(subset=['Documento'], inplace=True)  # Eliminar filas con valores NaN en 'Documento'

        # Contar la cantidad de documentos únicos y repetidos
        self.cantidad_registrados = datos_fil['Documento'].count()  # Cantidad de documentos registrados
        self.cantidad_repetidos = datos_fil.duplicated(subset=['Documento']).sum()  # Cantidad de documentos repetidos

        # Identificar y contar los documentos repetidos
        self.df_repetidos = datos_fil[datos_fil.duplicated(subset=['Documento'], keep=False)]
        self.conteo_repetidos = self.df_repetidos['Documento'].value_counts().reset_index()
        self.conteo_repetidos.columns = ['Documento', 'Cantidad']

    def obtener_resultados(self):
        # Devolver resultados
        resultados = {
            'Cantidad de Documentos Registrados:\n': self.cantidad_registrados,
            'Cantidad de Documentos Repetidos:\n': self.cantidad_repetidos,
            'Lista de Documentos Repetidos\n': self.conteo_repetidos
        }
        return resultados


# Uso de la clase
analisis = AnalisisCertificados(cert_1427)
resultados = analisis.obtener_resultados()

# Imprimir resultados
for clave, valor in resultados.items():
    print(f'{clave}: {valor}')
```

```python
import os
import pandas as pd
import warnings

class CertificadosProcessor:
    def __init__(self, pqr_file_path, reg_file_path):
        self.pqr_file_path = pqr_file_path
        self.reg_file_path = reg_file_path

    def read_pqr_data(self):
        try:
            cert_1427 = pd.read_csv(self.pqr_file_path, sep=';', encoding='ISO-8859-1', index_col=None, header=0)
        except FileNotFoundError:
            raise FileNotFoundError(f"No se encontró el archivo CSV en {self.pqr_file_path}.")
        
        return pd.DataFrame(cert_1427)

    def read_reg_data(self):
        try:
            reg_1427 = pd.read_excel(self.reg_file_path, sheet_name='CERTIFICADO_PQRS', header=0)
        except FileNotFoundError:
            print(f"Error: No se encontró el archivo Excel en {self.reg_file_path}.")
            reg_1427 = pd.DataFrame()

        return pd.DataFrame(reg_1427)

    def process_reg_data(self, df):
        # Convertir la columna 'FECHA_LIMITE' al formato de fecha
        df['FECHA_LIMITE'] = pd.to_datetime(df['FECHA_LIMITE'], errors='coerce')
        df['FECHA_DE_REMISION'] = pd.to_datetime(df['FECHA_DE_REMISION'], errors='coerce')
        # ... (resto de conversiones y formateos)

        return df

    def filter_reg_data(self, df):
        # Formatear los valores nulos y floats to int
        numeric_columns = ['DOCUMENTO', 'NUMERO_PQR', 'N_NCAPACIDAD_INICIAL', 'N_INCAPACIDAD_FINAL']
        for col in numeric_columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')
            df = df.dropna(subset=[col])
            df[col] = df[col].astype(int)

        return df

    def process_pqr_data(self, df):
        # Poner nombre a la columna de datos
        df.columns = ['Datos']
        # Dividir la columna 'Datos' en columnas separadas
        df[['TD', 'Documento', 'Nombre1', 'Nombre2', 'Nombre3', 'Apellido1', 'Apellido2', 'Apellido3']] = df['Datos'].str.split(' ', expand=True, n=7)

        return df

# Uso de la clase
pqr_datos = os.path.join(r"C:\Users\dapache\OneDrive - NUEVA EPS\CERTIFICADOS DECRECTO_1427_PQRS", "consolidado.txt")
reg_datos = os.path.join(r"C:\Users\dapache\OneDrive - NUEVA EPS\CERTIFICADOS DECRECTO_1427_PQRS", "certificados_1427_PQRS.xlsx")

processor = CertificadosProcessor(pqr_datos, reg_datos)

# Leer datos y procesar
pqr_data = processor.read_pqr_data()
reg_data = processor.read_reg_data()

processed_reg_data = processor.process_reg_data(reg_data)
filtered_reg_data = processor.filter_reg_data(processed_reg_data)

processed_pqr_data = processor.process_pqr_data(pqr_data)
```
